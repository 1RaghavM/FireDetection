```python
from ultralytics import YOLO
import cv2
from pathlib import Path
from torchvision import transforms
from PIL import Image
```

```python
image_dir = Path("data/train/images")
image_files = sorted(image_dir.glob("*.jpg"))

for img_path in image_files:
    img = cv2.imread(str(img_path))
    height, width = img.shape[:2]
    print(height, width)
```

```python
target_size = (640, 640)

directories = ['train', 'val', 'test']
for direc in directories:
    image_dir = Path("data", direc, "images")
    label_dir = Path("data", direc, "labels")
    
    image_files = sorted(image_dir.glob("*.jpg"))
    
    for img_path in image_files:
        img_cv2 = cv2.imread(str(img_path))
        original_height, original_width = img_cv2.shape[:2]
        
        img_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
        img_pil = Image.fromarray(img_rgb)
        
        # Resize image
        img_resized = img_pil.resize(target_size, Image.Resampling.LANCZOS)
        
        # Save transformed image
        img_resized.save(img_path)
        
        # Load and update corresponding label file
        label_path = label_dir / (img_path.stem + ".txt")
        if label_path.exists():
            # Read original labels
            with open(label_path, 'r') as f:
                lines = f.readlines()
            
            scale_x = target_size[0] / original_width
            scale_y = target_size[1] / original_height
            
            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) == 5:
                    class_id, x_center, y_center, width, height = map(float, parts)
                    
                    new_x_center = x_center  
                    new_y_center = y_center  
                    new_width = width  
                    new_height = height  
                    
                    updated_lines.append(f"{int(class_id)} {new_x_center} {new_y_center} {new_width} {new_height}\n")
            
            # Save updated labels
            with open(label_path, 'w') as f:
                f.writelines(updated_lines)
        
        print(f"Processed: {img_path.name}")
```

```python
model = YOLO("yolo11n")

model.train(
    data="data/dataset.yaml",
    epochs=100,
    imgsz=640,
    patience=20,
    save=True,
    workers=8,
    dropout=0.2,
    plots=True,
    batch=16,
    project="."
)
```
